```python
# Task 1: Write a function named capitalize_words that takes a string as input and returns a new string where every word is capitalized. Assume that words are separated by spaces.


# This is the function that you need to populate with your code
def capitalize_words(input_string):
    pass

    ## Write your code here
    words = input_string.split()
    capitalize_words = [word.capitalize() for word in words]
    return ' '.join(capitalize_words)


    ### Your code ends here
```

```python
# Task 3:

# This is the function that you need to populate with your code
def unique_elements(nested_lists):
    pass

    ## Write your code here
    combined_set = set(nested_lists[0])
    for lis in nested_lists[1:]:
        combined_set = combined_set.union(set(lis))
        union_list = sorted(combined_set)
    return union_list


    ### Your code ends here

    # Example tests for your reference
print(unique_elements([[1, 2, 3], [2, 3, 4], [4, 5, 6]]))
# Output: {1, 2, 3, 4, 5, 6}
print(unique_elements([[10, 20], [20, 30, 40], [10, 50]]))
# Output: {10, 20, 30, 40, 50}
```

Steps:
Convert the string to lowercase: This ensures the function is case-insensitive.
Remove punctuation: Strip out any punctuation to avoid counting words with punctuation marks as different words.
Split the string into words: Use the split() method to break the string into individual words.
Count the frequency of each word: Use a dictionary where the keys are words, and the values are their counts.

```python
# Task 4:

# This is the function that you need to populate with your code
def word_frequency(input_string):
    pass

    ## Write your code here
    result = " "
    for char in input_string:
        if char not in '''!()-[]{};:'"\,<>./?@#$%^&*_~''':
            result += char
    conv_str = result.lower()
    words = conv_str.split()
    freq = {}

    for word in words:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1
    return freq

    ### Your code ends here

    # Example tests for your reference
print(word_frequency("Hello world! Hello everyone."))
# Output: {'hello': 2, 'world': 1, 'everyone': 1}
print(word_frequency("Python programming is fun. Programming in Python is exciting."))
# Output: {'python': 2, 'programming': 2, 'is': 2, 'fun': 1, 'in': 1, 'exciting': 1}
```
