```python
# importing Numpy
import numpy as np
```

```python
# creating arrays
arr1 = np.array([1,2,3,4])
arr2 = np.array([[1,2,], [3,4], [5,6]])
```

```python
# Inspecting the arrays
print("Array 1:\n", arr1)
print("Array 2:\n", arr2)
```

```python
# Array properties
print("Shape of arr2:", arr2.shape)
print("Data type of arr2:", arr2.dtype)
print("Shape of arr1:", arr1.shape)
print("Data type of arr1:", arr1.dtype)
```

```python
# Slicing and indexing
print("First element of arr1:", arr1[0])
print("Firstrow of arr2:\n", arr2[0, :])
print("All elements in second column of arr2:", arr2[:, 1])
```

```python
# Reshaping arrays
arr3 = np.arange(1, 13).reshape(3, 4)
print("Reshaped array:\n", arr3)
```

```python
# performing mathematical operations
arr4 = np.array([1,2,3])
arr5 = np.array([4,5,6])
print("Element-wise addition:", arr4 + arr5)
print("Element-wise multiplication:", arr4 * arr5)
```

```python
# Broadcasting
arr6 = np.array([1,2,3])
print("Broadcasting example:\n", arr6 + 5)
```

```python
# Aggregate function
print("Sum of elements in arr6:", np.sum(arr6))
print("Mean of elements in arr6:", np.mean(arr6))
print("Max element in arr6:", np.max(arr6))
```
